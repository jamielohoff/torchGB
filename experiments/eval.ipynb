```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
font = {"family" : "normal",
        "weight" : "normal",
        "size"   : 15}

plt.rc("font", **font)

file = "oscar_de_to_en.csv" 
l = 80 # 2200

data = np.genfromtxt(os.path.join(os.getcwd(), file), delimiter=",", skip_header=1)
print(data.shape)
data
num_samples = data[:l, 0]
genomic_bottleneck_data = data[:l, 1]
baseline = data[:l, 4]


fig, ax = plt.subplots()
ax.loglog(num_samples, genomic_bottleneck_data, label="Training with GNet init")
ax.loglog(num_samples, baseline, label="Baseline")
ax.set_xlabel("# batches")
ax.set_ylabel("perplexity")

ax.set_title("Oscar-2301 dataset 'en' with pretraining on 'de'")
ax.legend(loc="upper right")
```
